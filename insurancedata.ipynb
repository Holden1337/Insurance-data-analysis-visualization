# Using Pandas and seaborn to analyze sample data regarding insurance prices

```python
#import necessary libraries
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#path to file
path = '/home/holden/programming/python/data_analysis/insurance.csv'

#read data
data = pd.read_csv(path)
```

```python
data.head()
```

## Create scatter plot to visualize correlation between BMI and monthly insurance cost

```python


#size
plt.figure(figsize=(15,6))

#title
plt.title("insurance costs vs BMI")


#creat scatter for bmi
sns.regplot(x=data['bmi'],y=data['charges'])
```

## Here we will add a third desciptor (smoker or non smoker) to see how this effects the data along with BMI 

```python
#size
plt.figure(figsize=(20,10))

sns.scatterplot(x=data['bmi'], y=data['charges'], hue=data['smoker'])
```

## Create a plot that has two different regression lines. One for smokers and one for Non-smokers

```python
#size
plt.figure(figsize=(40,10))

sns.lmplot(x="bmi", y="charges", hue="smoker", data=data)
```

## Build a categorical scatter plot to see a different visualization between smokers and non-smokers

```python


sns.swarmplot(x=data['smoker'],
              y=data['charges'])
```
